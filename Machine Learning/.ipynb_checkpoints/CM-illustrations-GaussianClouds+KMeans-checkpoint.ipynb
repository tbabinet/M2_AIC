## A look a some data: Gaussians (propto I, diagonal, full)


```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
m1 = np.zeros(2)
```

```python
cov1 = np.array( [[1.0,0],\
                  [0.,1]])
data1= np.random.multivariate_normal(m1,cov1,1500)
plt.figure()
plt.plot(data1[:,0], data1[:,1],'r+')
plt.xlim([-10,10])
plt.ylim([-10,10])
print(cov1)

cov1 = np.array( [[1.0,0],\
                  [0.,10.0]])
data1= np.random.multivariate_normal(m1,cov1,1500)
plt.figure()
plt.plot(data1[:,0], data1[:,1],'r+')
plt.xlim([-10,10])
plt.ylim([-10,10])
print(cov1)

cov1 = np.array( [[1.0,0.4],\
                  [0.4,1]])
data1= np.random.multivariate_normal(m1,cov1,1500)
plt.figure()
plt.plot(data1[:,0], data1[:,1],'r+')
plt.xlim([-10,10])
plt.ylim([-10,10])
print(cov1)

cov1 = np.array( [[1.0,-0.9],\
                  [-0.9,1]])
data1= np.random.multivariate_normal(m1,cov1,1500)
plt.figure()
plt.plot(data1[:,0], data1[:,1],'r+')
plt.xlim([-10,10])
plt.ylim([-10,10])
print(cov1)
```

```python
K=3
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=K, random_state=0)
affk = kmeans.fit_predict(alldata)
colors=["blue", "red", "green", "purple"]
for k in range(K):
    line = affk==k
    plt.plot(alldata[line,0],alldata[line,1], marker="+", linestyle="", color=colors[k])
```

```python
K=6
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=K, random_state=0, max_iter=1)
affk = kmeans.fit_predict(alldata)
colors=["blue", "red", "green", "purple","yellow", "pink"]
for k in range(K):
    line = affk==k
    plt.plot(alldata[line,0],alldata[line,1], marker="+", linestyle="", color=colors[k])
```
