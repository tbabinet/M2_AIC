```python
import requests
import selenium
from selenium import webdriver
import urllib.request
import time
import os
from bs4 import BeautifulSoup
```

```python
####Scrapping American Dad
url = 'https://www.springfieldspringfield.co.uk/episode_scripts.php?tv-show=american-dad'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
list_s = soup.find_all("div", "season-episodes")
i=1
for s in list_s:
    j=1
    for a in s.find_all('a'):
        name = a.get('href').split("=")[2]
        url_ep = "https://www.springfieldspringfield.co.uk/{}".format(a.get('href'))
        response = requests.get(url_ep)
        soup_ep = BeautifulSoup(response.text, "html.parser")
        script = soup_ep.find('div', "scrolling-script-container")
        with open(name, "w+") as f:
            f.write(script.text)
```

```python
####Scrapping artistes rap
album_list = []

dic = {"Pnl":["Deux-freres", "Dans-la-legende", "Le-monde-chico", "Que-la-famille"],
      "Sch": ["Jvlivs-tome-ii", "Jvlivs", "Deo-favente", "Anarchie", "A7"],
       "Jul": ["Rien-100-rien", "Album-gratuit-vol-5", "Rien-100-rien-reedition", "La-zone-en-personne", "Inspi-d-ailleurs", 
              "La-tete-dans-les-nuages", "Album-gratuit-vol-4", "Je-ne-me-vois-pas-briller", "Cassage-de-nuques", "Album-gratuit-vol-3",
             "L-ovni", "Album-gratuit-vol-2", "Emotions", "Album-gratuit-vol-1", "My-world-reedition-collector","Je-tourne-en-rond",
             "Album-gratuit-2015", "Je-trouve-pas-le-sommeil", "Lacrizeomic", "Dans-ma-paranoia"],
       "Heuss-lenfoire" : ["En-esprit"]
      }

for artiste in dic:
    url_artiste =  'https://genius.com/albums/{}/'.format(artiste)
    if not os.path.exists("{}".format(artiste)):
            os.mkdir("{}".format(artiste))
    for album in dic[artiste] :
        url_album = '{}{}'.format(url_artiste, album)
        response = requests.get(url_album)
        soup = BeautifulSoup(response.text, "html.parser")
        div = soup.find_all("div", "column_layout-column_span column_layout-column_span--primary")
        list_a = div[1].find_all("a", "u-display_block")
        
        directory = "{}/{}/".format(artiste, album)
        s = 1
        if not os.path.exists(directory):
            os.mkdir(directory)
            for a in list_a:
                url_paroles = a.get("href")
                response = requests.get(url_paroles)
                soup_p = BeautifulSoup(response.text, "html.parser")
                lyrics = soup_p.find("div", "lyrics")
                nf = directory+str(s)+".txt"
                s+=1
                with open(nf,"w+") as f:
                    try:
                        f.write(lyrics.text.strip())
                        print("{} OK".format(nf))
                    except:
                        print("{} pas OK".format(nf))
            print(album + " OK")
print("DONE")
```
