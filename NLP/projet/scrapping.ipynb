```python
import requests
import selenium
from selenium import webdriver
import urllib.request
import time
import os
from bs4 import BeautifulSoup, element
```

```python
####Scrapping American Dad
url = 'https://www.springfieldspringfield.co.uk/episode_scripts.php?tv-show=american-dad'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
list_s = soup.find_all("div", "season-episodes") ##liste des saisons
i=0
for s in list_s:
    i+=1
    s_name = "scripts_american_dad/s{}".format(i)
    os.mkdir(s_name)
    j=0
    for a in s.find_all('a'):
        j+=1
        filename = s_name+"/"+a.get('href').split("=")[2]+".txt"
        url_ep = "https://www.springfieldspringfield.co.uk/{}".format(a.get('href'))
        response = requests.get(url_ep)
        soup_ep = BeautifulSoup(response.text, "html.parser")
        script = soup_ep.find('div', "scrolling-script-container")
        txt  =""
        for c in script.children:
            if isinstance(c, element.NavigableString ):
                txt+=c.string.strip()+"\n"
        with open(filename, "wb+") as f:
            uni = u'{}'.format(txt)
            f.write(uni.encode('utf8'))
```
