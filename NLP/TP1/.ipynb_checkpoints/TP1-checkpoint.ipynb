```python
f = open("tbbts03e01.txt", "r")
txt=f.read()
lemmatizer = WordNetLemmatizer()
# dependency_parser = StanfordDependencyParser()
```

```python
word_tokens = nltk.word_tokenize(txt)
sent_tokens = nltk.sent_tokenize(txt)
pos_tokens = nltk.pos_tag(txt)
lemmatized_tokens = [lemmatizer.lemmatize(w) for w in word_tokens]
ne_tokens = nltk.ne_chunk(nltk.pos_tag(word_tokens))
```

```python
print(word_tokens[0:10])
print(sent_tokens[0:2])
print(pos_tokens[0:5])
print(lemmatized_tokens[0:20])
print(ne_tokens)
```
