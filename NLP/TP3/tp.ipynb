```python
import sklearn
import ri_basic
import eval.eval as eval
import os 
import nltk
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem import PorterStemmer
import operator
import matplotlib.pyplot as plt
from collections import Counter
```

```python
tk = []
for root,dirs,files in os.walk("data/texts"):
    for f in files:
        tk.append(ri_basic.get_tokens("data/texts/{}".format(f)))
tk = [w for txt in tk for w in txt]
```

```python
most_com_tk = ri_basic.get_most_common_tokens(tk, 20)
print(most_com_tk)
#Ce sont donc de très mauvais descripteurs
```

```python
lemmatizer = WordNetLemmatizer()
lems = [lemmatizer.lemmatize(w) for w in tk]
most_com_lems = ri_basic.get_most_common_tokens(lems, 20)
print(most_com_lems)
```

```python
stemmer = PorterStemmer()
stems = [stemmer.stem(w) for w in tk]
most_com_stems = ri_basic.get_most_common_tokens(stems, 20)
print(most_com_stems)
```

```python
pos_tags = nltk.pos_tag(tk)
most_com_tags = ri_basic.get_most_common_tokens(pos_tags, 15)
print(most_com_tags)
```

```python
d = dict()
for w, tag in pos_tags:
    if tag in d:
        d[tag]=d[tag]+1
    else:
        d[tag] = 0
sorted_d = sorted(d.items(), key=operator.itemgetter(1))[::-1]
print(sorted_d)
#On observe que la catégorie la plus fréquente est celle des noms. C'est une bonne chose car cela permet d'obtenir
#beaucoup d'informations sur un texte
#Pour obtenir de bons descripteurs, il faudrait garder les noms, les verbes
```

```python
#EXERCICE 2
def create_tfidf(dir):
    text_list = []
    text_name = []
    for subdir, dirs, files in os.walk(dir):
        for file in files:
            if file.endswith(".txt"):
                #print("treating "+file)
                file_path = subdir + os.path.sep + file
                #text_dict[file] = get_text(file_path)
                text_list.append(ri_basic.get_text(file_path))
                text_name.append(file)
    return text_list, text_name

tfidf = create_tfidf("data/texts")
```
